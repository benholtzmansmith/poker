```python
total_hands_count = sum([y for (x,y) in hands_with_count])
```

```python
def getRangeByPercent(frequency):
    overall_count = 0
    hands = []
    hc = float(total_hands_count)
    for (hand, count) in hands_with_count:
        overall_count = overall_count+count
        hands.append(hand)
        if(frequency < overall_count/hc):
            return hands
            
```

```python
ran = getRangeByPercent(.03)
','.join(ran)
```

```python
def calculateThreeBetFraction(o3bCallF, o4BetF):
    print("running")
    raiseSize = 11
    openSize = 3
    bb = 1

    foldF = 1.0 - o3bCallF - o4BetF  
    
    #they don't open 100% of hands so this isn't the best assumption
    opponent_four_bet_hands = getRangeByPercent(o4BetF)
    opponent_three_bet_call_hands = getRangeByPercent(o3bCallF)
    opponent_four_bet_hand_range = eval7.HandRange(', '.join(opponent_four_bet_hands))
    opponent_three_bet_call_hand_range = eval7.HandRange(', '.join(opponent_three_bet_call_hands))
    
    for i in reversed(range(len(all_hands))):
        hands = all_hands[0:i]
        hero_range = eval7.HandRange(', '.join(hands))
        _equities = equity.py_all_hands_vs_range(
            hero_range, 
            opponent_four_bet_hand_range, 
            eval7.HandRange(""), 
            1000
        )
        equities_l = [x for x in _equities.values()]
        _equity = sum(equities_l)/len(equities_l)
        evOnCall = o3bCallF * (raiseSize*2* _equity - raiseSize*2*(1 - _equity))
        evOnFold = foldF * (bb + openSize)
        evOn4bet = -1 * o4BetF * raiseSize
        totalEv = evOnCall + evOnFold + evOn4bet
        if(totalEv > 0):
            print("optimal range: ")
            print(hands)
            return (hero_range)
        else:
            print (totalEv, hands)
```
